```python
import mysql.connector
mydb = mysql.connector.connect(
 host="localhost",
 user="root",
 password="Venkatraj@02",
 database="agridata"
)
print(mydb)
mycursor = mydb.cursor(buffered=True)
```

```python
mycursor.execute("CREATE DATABASE agridata")
```

```python
#Creating a SQL Table
columns = agri.columns
col_defs = []
for col in columns:
    # Basic type mapping: float64 -> FLOAT, int64 -> INT, object -> VARCHAR(255)
    dtype = agri[col].dtype
    if dtype == 'float64':
        sql_type = 'FLOAT'
    elif dtype == 'int64':
        sql_type = 'INT'
    else:
        sql_type = 'VARCHAR(255)'
    # Replace spaces and parentheses for SQL compatibility
    colname = col.replace(' ', '_').replace('(', '').replace(')', '')
    col_defs.append(f"`{colname}` {sql_type}")

table_def = ",\n  ".join(col_defs)
create_table_sql = f"""
CREATE TABLE agridata.agri (
  {table_def}
);
"""

mycursor.execute(create_table_sql)
```

```python
mydb.commit()
```

```python
# Insert data from agri DataFrame into MySQL table
for _, row in agri.iterrows():
    values = []
    for col in columns:
        val = row[col]
        # Convert NaN to None for SQL NULL
        if pd.isna(val):
            values.append(None)
        else:
            values.append(val)
    placeholders = ','.join(['%s'] * len(values))
    insert_sql = f"INSERT INTO agridata.agri VALUES ({placeholders})"
    mycursor.execute(insert_sql, values)

mydb.commit()
```

```python
# Rice Production Across States by Year: Top 3 States Each Year
sql = """
SELECT 
    t.Year,
    t.State_Name,
    t.total_rice_production
FROM (
    SELECT 
        `Year`,
        `State_Name`,
        SUM(`RICE_PRODUCTION_1000_tons`) AS total_rice_production,
        RANK() OVER (PARTITION BY `Year` ORDER BY SUM(`RICE_PRODUCTION_1000_tons`) DESC) as rnk
    FROM agridata.agri
    GROUP BY `Year`, `State_Name`
) t
WHERE t.rnk <= 3
ORDER BY t.Year, t.rnk;
"""
mycursor.execute(sql)
out = mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
#Top 5 Districts by Wheat Yield Increase Over the Last 5 Years
sql = """
SELECT 
    `Dist_Name`,
    (MAX(`WHEAT_YIELD_Kg_per_ha`) - MIN(`WHEAT_YIELD_Kg_per_ha`)) AS yield_increase
FROM agridata.agri
WHERE `Year` >= (SELECT MAX(`Year`) FROM agridata.agri) - 4
GROUP BY `Dist_Name`
ORDER BY yield_increase DESC
LIMIT 5;
"""
mycursor.execute(sql)
out=mycursor.fetchall() #fectchhall it will fetch all the data from the file
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#States with the Highest Growth in Oilseed Production (5-Year Growth Rate)
sql = """
SELECT 
    `State_Name`,
    (SUM(CASE WHEN `Year` = (SELECT MAX(`Year`) FROM agridata.agri) THEN `OILSEEDS_PRODUCTION_1000_tons` ELSE 0 END) -
     SUM(CASE WHEN `Year` = (SELECT MAX(`Year`) FROM agridata.agri) - 5 THEN `OILSEEDS_PRODUCTION_1000_tons` ELSE 0 END)
    ) / NULLIF(SUM(CASE WHEN `Year` = (SELECT MAX(`Year`) FROM agridata.agri) - 5 THEN `OILSEEDS_PRODUCTION_1000_tons` ELSE 0 END), 0) AS growth_rate
FROM agridata.agri
WHERE `Year` IN ((SELECT MAX(`Year`) FROM agridata.agri), (SELECT MAX(`Year`) FROM agridata.agri) - 5)
GROUP BY `State_Name`
ORDER BY growth_rate DESC
LIMIT 5;
"""
mycursor.execute(sql)
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)
sql = """
SELECT
    `Dist_Name`,
    (
        (COUNT(*) * SUM(`RICE_AREA_1000_ha` * `RICE_PRODUCTION_1000_tons`) - SUM(`RICE_AREA_1000_ha`) * SUM(`RICE_PRODUCTION_1000_tons`)) /
        (SQRT(COUNT(*) * SUM(POW(`RICE_AREA_1000_ha`, 2)) - POW(SUM(`RICE_AREA_1000_ha`), 2)) *
         SQRT(COUNT(*) * SUM(POW(`RICE_PRODUCTION_1000_tons`, 2)) - POW(SUM(`RICE_PRODUCTION_1000_tons`), 2)))
    ) AS rice_corr,
    (
        (COUNT(*) * SUM(`WHEAT_AREA_1000_ha` * `WHEAT_PRODUCTION_1000_tons`) - SUM(`WHEAT_AREA_1000_ha`) * SUM(`WHEAT_PRODUCTION_1000_tons`)) /
        (SQRT(COUNT(*) * SUM(POW(`WHEAT_AREA_1000_ha`, 2)) - POW(SUM(`WHEAT_AREA_1000_ha`), 2)) *
         SQRT(COUNT(*) * SUM(POW(`WHEAT_PRODUCTION_1000_tons`, 2)) - POW(SUM(`WHEAT_PRODUCTION_1000_tons`), 2)))
    ) AS wheat_corr,
    (
        (COUNT(*) * SUM(`MAIZE_AREA_1000_ha` * `MAIZE_PRODUCTION_1000_tons`) - SUM(`MAIZE_AREA_1000_ha`) * SUM(`MAIZE_PRODUCTION_1000_tons`)) /
        (SQRT(COUNT(*) * SUM(POW(`MAIZE_AREA_1000_ha`, 2)) - POW(SUM(`MAIZE_AREA_1000_ha`), 2)) *
         SQRT(COUNT(*) * SUM(POW(`MAIZE_PRODUCTION_1000_tons`, 2)) - POW(SUM(`MAIZE_PRODUCTION_1000_tons`), 2)))
    ) AS maize_corr
FROM agridata.agri
GROUP BY `Dist_Name`
ORDER BY `Dist_Name`;
"""
mycursor.execute(sql)
out = mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
# Yearly Production Growth of Cotton in Top 5 Cotton Producing States
sql = """
SELECT 
    `Year`,
    `State_Name`,
    SUM(`COTTON_PRODUCTION_1000_tons`) AS total_cotton_production
FROM agridata.agri
WHERE `State_Name` IN (
    SELECT `State_Name`
    FROM (
        SELECT `State_Name`, SUM(`COTTON_PRODUCTION_1000_tons`) AS total_production
        FROM agridata.agri
        GROUP BY `State_Name`
        ORDER BY total_production DESC
        LIMIT 5
    ) AS top_states
)
GROUP BY `Year`, `State_Name`
ORDER BY `State_Name`, `Year`;
"""
mycursor.execute(sql)
out = mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```
